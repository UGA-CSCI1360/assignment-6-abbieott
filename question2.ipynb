```python
# You solution goes here
# you are encouraged to use multiple cells in the notebook

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

titanic_data = pd.read_csv('train.csv')
print(titanic_data['Embarked'].value_counts())

survivors = titanic_data[titanic_data['Survived'] == 1]
casualties = titanic_data[titanic_data['Survived'] == 0]

#age and fare for both survivors and casualties
plt.figure(figsize=(12, 6))
plt.scatter(survivors['Fare'], survivors['Age'], label='Survivors', alpha=0.7, marker='o')
plt.scatter(casualties['Fare'], casualties['Age'], label='Casualties', alpha=0.7, marker='x')
plt.title('Scatter Plot of Fare vs Age for Survivors and Casualties')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.legend()
plt.show()

# survivors by port
survived_embarked_counts = survivors['Embarked'].value_counts()
plt.bar(survived_embarked_counts.index, survived_embarked_counts.values, alpha=0.5, color = 'green', label='Survived')
plt.title('Histogram of Embarked for Survivals')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.legend()
plt.show()

# Casualties by port
casualties_embarked_counts = casualties['Embarked'].value_counts()
plt.bar(casualties_embarked_counts.index, casualties_embarked_counts.values, alpha=0.5, label='Casualties')
plt.title('Histogram of Embarked for Casualties')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.legend()
plt.show()
```
